```python
# EDA on Cardiotocographic Dataset

# ðŸ“¦ Step 1: Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns



# ðŸ“¥ Step 2: Load the Dataset
df = pd.read_csv("Cardiotocographic.csv")  # Make sure this file is in the same directory
df.head()
```

### ðŸ§¹ Step 3: Data Cleaning & Preparation

We'll now check for missing values, data types, and clean the dataset if necessary.


```python
# Check for missing values
df.isnull().sum()
```

```python
# Convert columns to numeric (in case there are some wrong types)
df = df.apply(pd.to_numeric, errors='coerce')

# Drop rows with NaN (after coercion)
df.dropna(inplace=True)

# Final dataset shape
df.shape
```

### ðŸ§® Step 4: Statistical Summary

Let's now see the basic statistics of each variable â€” mean, median, std, etc.


```python
# Statistical description
df.describe()
```

### ðŸ“Š Step 5: Data Visualization â€“ Histograms

We'll now plot histograms to see distribution of each numerical feature.


```python
# Plot histograms
df.hist(figsize=(15, 12), bins=20, color='skyblue', edgecolor='black')
plt.suptitle("Histograms of All Features", fontsize=18)
plt.tight_layout()
plt.show()
```

### ðŸ“¦ Step 6: Outlier Detection â€“ Boxplots


```python
# Boxplots to detect outliers
plt.figure(figsize=(15, 10))
sns.boxplot(data=df)
plt.title("Boxplots of All Features")
plt.xticks(rotation=90)
plt.show()
```

### ðŸ”¥ Step 7: Correlation Analysis â€“ Heatmap


```python
# Heatmap of correlations
plt.figure(figsize=(14, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

### ðŸ§ª Step 8: Pairplot for Visual Patterns


```python
# Pairplot (only first 5 columns for simplicity)
sns.pairplot(df[df.columns[:5]])
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()
```

### ðŸŽ» Step 9: Violin Plot


```python
# Violin plot for selected columns
plt.figure(figsize=(10, 6))
sns.violinplot(data=df[['LB', 'AC', 'FM']])
plt.title("Violin Plot of LB, AC, FM")
plt.show()
```

### ðŸ’¡ Step 10: Insights

- LB (Baseline Heart Rate) is approximately normally distributed.
- Some features like `UC`, `DL`, and `DP` show outliers.
- Strong correlation between `ASTV` and `ALTV`.
- Dataset is clean and ready for modeling.
