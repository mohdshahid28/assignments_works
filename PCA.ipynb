```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import numpy as np

X = df[num_cols].values
scaler = StandardScaler()
X_std = scaler.fit_transform(X)

pca_full = PCA()
X_pca_full = pca_full.fit_transform(X_std)

explained = pca_full.explained_variance_ratio_
cum_explained = np.cumsum(explained)

```

```python

# Scree plot
plt.figure()
plt.plot(np.arange(1, len(explained)+1), explained, marker='o')
plt.xlabel("Principal Component")
plt.ylabel("Explained Variance Ratio")
plt.title("Scree Plot")
plt.show()
```

```python

# Cumulative explained variance
plt.figure()
plt.plot(np.arange(1, len(cum_explained)+1), cum_explained, marker='o')
plt.axhline(0.90, linestyle='--')
plt.axhline(0.95, linestyle='--')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Cumulative Explained Variance")
plt.show()
```

```python




# Choose components to reach ≥95% variance
n_opt = int(np.argmax(cum_explained >= 0.95) + 1) if len(cum_explained) else 1
print(f"Selected n_components = {n_opt} (≥95% variance)")

pca = PCA(n_components=n_opt)
X_pca = pca.fit_transform(X_std)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score

def evaluate_kmeans(X_data, k_range=range(2, 11), random_state=42):
    inertia_vals, sil_scores, db_scores = [], [], []
    models = {}
    for k in k_range:
        km = KMeans(n_clusters=k, n_init=10, random_state=random_state)
        labels = km.fit_predict(X_data)
        inertia_vals.append(km.inertia_)
        sil_scores.append(silhouette_score(X_data, labels))
        db_scores.append(davies_bouldin_score(X_data, labels))
        models[k] = (km, labels)
    return {
        'k_range': list(k_range),
        'inertia': inertia_vals,
        'silhouette': sil_scores,
        'davies_bouldin': db_scores,
        'models': models
    }

def best_k_by_silhouette(results):
    i = int(np.argmax(results['silhouette']))
    return results['k_range'][i], results['silhouette'][i]
```

```python
# Evaluate
res_orig = evaluate_kmeans(X_std, k_range=range(2, 11))
# Elbow (inertia)
plt.figure()
plt.plot(res_orig['k_range'], res_orig['inertia'], marker='o')
plt.xlabel("k"); plt.ylabel("Inertia"); plt.title("Elbow — Original (Std.)")
plt.show()
```

```python
# Silhouette
plt.figure()
plt.plot(res_orig['k_range'], res_orig['silhouette'], marker='o')
plt.xlabel("k"); plt.ylabel("Silhouette"); plt.title("Silhouette — Original (Std.)")
plt.show()
```

```python
# Davies–Bouldin
plt.figure()
plt.plot(res_orig['k_range'], res_orig['davies_bouldin'], marker='o')
plt.xlabel("k"); plt.ylabel("Davies–Bouldin"); plt.title("Davies–Bouldin — Original (Std.)")
plt.show()
```

```python
best_k_o, best_sil_o = best_k_by_silhouette(res_orig)
km_o, labels_o = res_orig['models'][best_k_o]
db_o = davies_bouldin_score(X_std, labels_o)
print(f"Best k (Original): {best_k_o} | Silhouette={best_sil_o:.3f} | DB={db_o:.3f}")
```

```python
pc2_orig = pca_full.transform(X_std)[:, :2]
plt.figure()
plt.scatter(pc2_orig[:,0], pc2_orig[:,1], c=labels_o)
plt.xlabel("PC1"); plt.ylabel("PC2")
plt.title(f"k={best_k_o} — Clusters on Original Data (shown in PCA 2D)")
plt.show()
```

```python
res_pca = evaluate_kmeans(X_pca, k_range=range(2, 11))
# Elbow
plt.figure()
plt.plot(res_pca['k_range'], res_pca['inertia'], marker='o')
plt.xlabel("k"); plt.ylabel("Inertia"); plt.title("Elbow — PCA Data")
plt.show()
```

```python
# Silhouette
plt.figure()
plt.plot(res_pca['k_range'], res_pca['silhouette'], marker='o')
plt.xlabel("k"); plt.ylabel("Silhouette"); plt.title("Silhouette — PCA Data")
plt.show()
```

```python

# Davies–Bouldin
plt.figure()
plt.plot(res_pca['k_range'], res_pca['davies_bouldin'], marker='o')
plt.xlabel("k"); plt.ylabel("Davies–Bouldin"); plt.title("Davies–Bouldin — PCA Data")
plt.show()
```

```python

best_k_p, best_sil_p = best_k_by_silhouette(res_pca)
km_p, labels_p = res_pca['models'][best_k_p]
db_p = davies_bouldin_score(X_pca, labels_p)
print(f"Best k (PCA): {best_k_p} | Silhouette={best_sil_p:.3f} | DB={db_p:.3f}")
```

```python
if X_pca.shape[1] >= 2:
    plt.figure()
    plt.scatter(X_pca[:,0], X_pca[:,1], c=labels_p)
    plt.xlabel("PC1"); plt.ylabel("PC2")
    plt.title(f"k={best_k_p} — Clusters on PCA Data (2D)")
    plt.show()
else:
    print("Only one PCA component selected; 2D plot not possible.")
```

```python
import pandas as pd

comparison = pd.DataFrame({
    "Setting": ["Original (Std.)", "PCA"],
    "Best k (Silhouette)": [best_k_o, best_k_p],
    "Best Silhouette": [best_sil_o, best_sil_p],
    "Best Davies–Bouldin": [db_o, db_p],
    "PCA n_components": [np.nan, X_pca.shape[1]]
})
display(comparison)

notes = []
if best_sil_p > best_sil_o:
    notes.append("PCA ↑ silhouette → better separation.")
elif best_sil_p < best_sil_o:
    notes.append("Original ↑ silhouette → PCA didn’t help here.")
else:
    notes.append("Similar silhouette on both.")

if db_p < db_o:
    notes.append("PCA ↓ Davies–Bouldin → tighter/more distinct clusters.")
elif db_p > db_o:
    notes.append("Original ↓ Davies–Bouldin → clusters more distinct without PCA.")
else:
    notes.append("Similar Davies–Bouldin on both.")

cum_95 = np.cumsum(pca_full.explained_variance_ratio_)[X_pca.shape[1]-1]*100 if X_pca.shape[1] >= 1 else 0
notes.append(f"PCA retained ≈{cum_95:.2f}% variance using {X_pca.shape[1]} components.")
print("\n".join("• " + m for m in notes))
```
